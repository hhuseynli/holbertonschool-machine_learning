## 0. Line Plot

**File:** `0-line.py`

````python
import numpy as np
import matplotlib.pyplot as plt

def line():
    y = np.arange(0, 11) ** 3
    plt.figure(figsize=(6.4, 4.8))
    plt.xlim(0, 10)
    plt.plot(y, "r")
    plt.show()
````

**Key points:**
- `plt.figure(figsize=(w, h))` — create a figure with specific dimensions (inches)
- `plt.xlim(min, max)` — set x-axis limits
- `plt.plot(y, 'r')` — line plot; `'r'` sets color to red (short color code)
- `plt.show()` — display the plot

## 1. Scatter Plot

**File:** `1-scatter.py`

````python
def scatter():
    mean = [69, 0]
    cov = [[15, 8], [8, 15]]
    np.random.seed(5)
    x, y = np.random.multivariate_normal(mean, cov, 2000).T
    y += 180
    plt.figure(figsize=(6.4, 4.8))
    plt.xlabel("Height (in)")
    plt.ylabel("Weight (lbs)")
    plt.title("Men's Height vs Weight")
    plt.scatter(x, y, c="magenta")
````

**Key points:**
- `np.random.multivariate_normal(mean, cov, n)` — generate n samples from a multivariate normal distribution
- `plt.scatter(x, y, c=color)` — scatter plot
- `plt.xlabel(label)`, `plt.ylabel(label)`, `plt.title(label)` — axis labels and title

## 2. Logarithmic Scale

**File:** `2-change_scale.py`

````python
def change_scale():
    x = np.arange(0, 28651, 5730)
    r = np.log(0.5)
    t = 5730
    y = np.exp((r / t) * x)
    plt.figure(figsize=(6.4, 4.8))
    plt.xlabel("Time (years)")
    plt.ylabel("Fraction Remaining")
    plt.title("Exponential Decay of C-14")
    plt.yscale("log")
    plt.xlim(0, 28650)
    plt.plot(x, y)
````

**Key points:**
- `plt.yscale('log')` — set y-axis to logarithmic scale
- Useful for exponential decay/growth data

## 3. Multiple Lines on One Plot

**File:** `3-two.py`

````python
def two():
    x = np.arange(0, 21000, 1000)
    r = np.log(0.5)
    t1 = 5730
    t2 = 1600
    y1 = np.exp((r / t1) * x)
    y2 = np.exp((r / t2) * x)
    plt.figure(figsize=(6.4, 4.8))
    plt.xlabel("Time (years)")
    plt.ylabel("Fraction Remaining")
    plt.title("Exponential Decay of Radioactive Elements")
    plt.xlim(0, 20000)
    plt.ylim(0, 1)
    plt.plot(x, y1, "r--", label="C-14")
    plt.plot(x, y2, "g", label="Ra-226")
    plt.legend(loc="upper right")
````

**Key points:**
- Multiple `plt.plot()` calls draw on the same axes
- `'r--'` — red dashed line, `'g'` — green solid line
- `label="..."` — assign labels for legend
- `plt.legend(loc=...)` — show legend at a location (`'upper right'`, `'lower left'`, etc.)
- `plt.ylim(min, max)` — set y-axis limits

## 4. Histogram

**File:** `4-frequency.py`

````python
def frequency():
    np.random.seed(5)
    student_grades = np.random.normal(68, 15, 50)
    plt.figure(figsize=(6.4, 4.8))
    plt.ylim(0, 30)
    plt.xlim(0, 100)
    plt.xticks(np.arange(0, 101, 10))
    plt.yticks(np.arange(0, 31, 5))
    plt.title("Project A")
    plt.xlabel("Grades")
    plt.ylabel("Number of Students")
    bins = np.arange(0, 101, 10)
    plt.hist(student_grades, bins=bins, edgecolor="black")
````

**Key points:**
- `np.random.normal(mean, std, size)` — generate normally distributed samples
- `plt.hist(data, bins=bins, edgecolor='black')` — histogram
- `plt.xticks(ticks)`, `plt.yticks(ticks)` — set specific tick positions

## 5. Stacked Bar Chart

**File:** `6-bars.py`

````python
def bars():
    np.random.seed(5)
    fruit = np.random.randint(0, 20, (4, 3))
    categories = ["Farrah", "Fred", "Felicia"]
    level1 = fruit[0]
    level2 = fruit[1]
    level3 = fruit[2]
    level4 = fruit[3]
    plt.figure(figsize=(6.4, 4.8))
    plt.title("Number of Fruit per Person")
    plt.ylabel("Quantity of Fruit")
    plt.ylim(0, 80)
    plt.bar(categories, level1, color="red", width=0.5)
    plt.bar(categories, level2, bottom=level1, color="yellow", width=0.5)
    plt.bar(categories, level3, bottom=level1+level2, color="#ff8000", width=0.5)
    plt.bar(categories, level4, bottom=level1+level2+level3, color="#ffe5b4", width=0.5)
    plt.legend(["apples", "bananas", "oranges", "peaches"])
````

**Key points:**
- `np.random.randint(low, high, shape)` — random integers
- `plt.bar(x, height, bottom=..., width=..., color=...)` — bar chart
- `bottom=` accumulates heights for stacking
- `width=0.5` — bar width (fraction of category spacing)
- `plt.legend([...])` — legend without explicit labels in the plotting calls

## 6. Scatter Plot with Color Gradient (Elevation)

**File:** `100-gradient.py`

````python
def gradient():
    np.random.seed(5)
    x = np.random.randn(2000) * 10
    y = np.random.randn(2000) * 10
    z = np.random.rand(2000) + 40 - np.sqrt(np.square(x) + np.square(y))
    plt.figure(figsize=(6.4, 4.8))
    plt.title("Mountain Elevation")
    plt.xlabel("x coordinate (m)")
    plt.ylabel("y coordinate (m)")
    plt.scatter(x, y, c=z)
    plt.colorbar(label="elevation (m)")
````

**Key points:**
- `plt.scatter(x, y, c=z)` — color each point by a third variable `z`
- `plt.colorbar(label=...)` — add a color bar with a label
- Color gradient is useful for visualizing elevation, density, or any continuous value
